```python
#import libs
import pandas as pd
import sqlite3
```

```python
#wrapper for pd.read.sql.query
def rsql(query):
    return pd.read_sql_query(query,conn)
```

```python
#create a connection to database
conn=sqlite3.connect('household.db')
#create a cursor
c=conn.cursor()
```

```python
#Lets read data and extract col names
expense=pd.read_csv('Income_expense_data.csv')
```

```python
#add data to table
expense.to_sql('expense',conn,if_exists = 'replace',index = False)
```

```python
#lets extract column names
print(expense.columns)
```

```python
c.execute('''
    CREATE TABLE IF NOT EXISTS trade(
        'Date' INT, 'Province' TEXT, 'Expenditures' TEXT, 'COORDINATE' TEXT, 'Expense' INT, 'Income' INT
    );
''')
```

```python
rsql("SELECT * FROM expense")
```

```python
rsql("SELECT * FROM expense WHERE Income > 100000")
```

```python
rsql("SELECT Province, SUM(Expense) FROM expense GROUP BY Province")
```

```python
rsql("SELECT * FROM expense ORDER BY Expense DESC")
```

```python
rsql("SELECT * FROM expense LIMIT 5")
```

```python
rsql("SELECT Income AS Total_Income, Expense AS Total_Expense FROM expense LIMIT 5")
```

```python
rsql("""
SELECT Province, AVG(Expense) AS Avg_Expense 
FROM expense 
GROUP BY Province 
HAVING Avg_Expense > 5000
""")
```

```python
rsql("SELECT * FROM expense WHERE Income BETWEEN 100000 AND 500000 LIMIT 5")
```

```python
rsql("""
SELECT 
    AVG(Expense) AS Avg_Expense,
    SUM(Income) AS Total_Income,
    COUNT(*) AS Total_Records
FROM expense
""")
```

```python
rsql("SELECT DISTINCT Province FROM expense")
```
